```python
import datetime
import time
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from pycoingecko import CoinGeckoAPI as cg
```

```python
cg = CoinGeckoAPI()
```

```python
data = cg.get_coin_market_chart_by_id(id='bitcoin',vs_currency='usd',days=5000)
unix_time = []
price = []
for i in data['prices']:
    unix_time.append(i[0])
    price.append(i[1])
print(unix_time)

human_time = []
for i in unix_time:
    human_time.append(datetime.datetime.fromtimestamp(i/1000))

```

```python

    
print(type(price[0]))
print(len(human_time))
print(len(price))

for i in range(0,len(human_time)):
    print(str(human_time[i]) + '\t' + str(price[i]))

```

```python
fig, ax = plt.subplots(figsize=(13, 6))
ax.title.set_text('Bitcoin')
ax.set_xlabel('Date')
ax.set_ylabel('USD')
ax.set_yscale('log')
ax.grid(visible=True, which='major', axis='both')
ax.plot(human_time, price)
```

```python
fig1, axs1 = plt.subplots(3, 1, figsize=(13, 13), layout='constrained')

# common to all three:
for ax1 in axs1:
    ax1.plot(human_time, price)
    
    # Major ticks every half year, minor ticks every month,
    ax1.xaxis.set_major_locator(mdates.MonthLocator(bymonth=(1, 7)))
    ax1.xaxis.set_minor_locator(mdates.MonthLocator())
    ax1.grid(True)
    ax1.set_ylabel(r'Price [\$]')

# different formats:
ax1 = axs1[0]
ax1.set_title('DefaultFormatter', loc='left', y=0.85, x=0.02, fontsize='medium')

ax1 = axs1[1]
ax1.set_title('ConciseFormatter', loc='left', y=0.85, x=0.02, fontsize='medium')
ax1.xaxis.set_major_formatter(mdates.ConciseDateFormatter(ax.xaxis.get_major_locator()))

ax1 = axs1[2]
ax1.set_title('Manual DateFormatter', loc='left', y=0.85, x=0.02, fontsize='medium')

# Text in the x-axis will be displayed in 'YYYY-mm' format.
#ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%b'))
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Rotates and right-aligns the x labels so they don't crowd each other.
for label in ax1.get_xticklabels(which='major'):
    label.set(rotation=30, horizontalalignment='right')

plt.show()
```
